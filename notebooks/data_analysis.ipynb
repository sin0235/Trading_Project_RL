```python
import pandas as pd
import numpy as np
import os
```

```python
dataset_path = "data/raw"
if not os.path.exists(dataset_path):
    dataset_path = os.path.join("..", "data", "raw")
csv_files = [f for f in os.listdir(dataset_path) if f.endswith('.csv')]
print(csv_files)
```

```python
dataset = []
for csv_file in csv_files:
    file_path = os.path.join(dataset_path, csv_file)
    data = pd.read_csv(file_path)
    dataset.append(data)
print(dataset[0].columns)
```

```python
start_times = []
end_times = []
for i in range(len(dataset)):
    start_times.append(dataset[i].iloc[0]['time'])
    end_times.append(dataset[i].iloc[-1]['time'])
print("Thời gian bắt đầu: ", start_times)
print("Thời gian kết thúc: ", end_times)
```

```python
for df in dataset:
    print(df.duplicated().sum())
```
